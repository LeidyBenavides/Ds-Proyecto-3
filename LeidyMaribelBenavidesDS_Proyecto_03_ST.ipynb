# Componentes de la Serie

El método más sencillo se conoce como media móvil. Consiste en tomar una ventana de tiempo suficientemente grande y promediar la serie sobre esa ventana.

```python
ventana = int(365*2/7)
diario_promedio_tendencia = diario.cantidad_pasos.rolling(window=ventana, center = True).mean()
diario_promedio_tendencia
```

```python
ventana1 = int(365*2/7)
diario_promedio_tendencia1 = diario1.cantidad_pasos.rolling(window=ventana1, center = True).mean()
diario_promedio_tendencia1
ventanae = int(365*2/7)
diario_promedio_tendenciae = diarioelim.cantidad_pasos.rolling(window=ventanae, center = True).mean()
diario_promedio_tendenciae
ventanaS = int(365*2/7)
diario_promedio_tendenciaS = diarioSal.cantidad_pasos.rolling(window=ventanae, center = True).mean()
diario_promedio_tendenciaS
```

### 2.1. One-Step Prediction

Vamos a entrenar un `RandomForestRegressor` para predecir el siguiente valor en la serie de tiempo diario del flujo vehicular por la autopista Illia. Utilizaremos los dos primeros años como conjunto de entrenamiento. Y el último como evaluación.

```python
senial = diario.cantidad_pasos
senial.head()
senial1 = diario1.cantidad_pasos

seniale = diarioelim.cantidad_pasos
senialS = diarioSal.cantidad_pasos
```

En primer lugar, debemos definir cuántos valores *hacia atrás* tomaremos.

```python
look_back = 52*2 # usamos dos anios de datos para predecir el valor siguiente
```

Creamos los `X` e `y` con ceros para luego llenar con los datos. 

```python
N = senial.shape[0]
X = np.zeros((N - look_back - 1,look_back))
y = np.zeros(N - look_back - 1)
print(X.shape, y.shape)
NS = senialS.shape[0]
Xs = np.zeros((NS - look_back - 1,look_back))
ys = np.zeros(NS - look_back - 1)
print(Xs.shape, ys.shape)
N1 = senial1.shape[0]
X1 = np.zeros((N1 - look_back - 1,look_back))
y1 = np.zeros(N1 - look_back - 1)
print(X1.shape, y1.shape)
Ne = seniale.shape[0]
Xe = np.zeros((Ne- look_back - 1,look_back))
ye = np.zeros(Ne - look_back - 1)
print(Xe.shape, ye.shape)
```

senial.shape[0]

Llenamos los `X` e `y`

```python
for i in range(X.shape[0]):
#     X[i,:] = senial.values.iloc[i:i+look_back]
#     y[i] = senial.values.iloc[i+look_back]
    X[i,:] = senial.iloc[i:i+look_back]
    y[i] = senial.iloc[i+look_back]
```

```python
for i in range(Xs.shape[0]):
#     X[i,:] = senial.values.iloc[i:i+look_back]
#     y[i] = senial.values.iloc[i+look_back]
    Xs[i,:] = senialS.iloc[i:i+look_back]
    ys[i] = senialS.iloc[i+look_back]
```

```python
for i in range(X1.shape[0]):
#     X[i,:] = senial.values.iloc[i:i+look_back]
#     y[i] = senial.values.iloc[i+look_back]
    X1[i,:] = senial1.iloc[i:i+look_back]
    y1[i] = senial1.iloc[i+look_back]

for i in range(Xe.shape[0]):
#     X[i,:] = senial.values.iloc[i:i+look_back]
#     y[i] = senial.values.iloc[i+look_back]
    Xe[i,:] = seniale.iloc[i:i+look_back]
    ye[i] = seniale.iloc[i+look_back]
```

Es importante notar que el tiempo de la medición - fecha - ya no es parte de los datos.

Hacemos el train/test split.

```python
N_train = 10*52 - look_back # cantidad de instancias que vamos a usar para entrenar. 
N_test = 4*52 #cantidad de puntos que vamos a usar para evaluar
N_total = N_train + N_test

length_total = N_train + N_test + look_back

X_train = X[:N_train,:]
y_train = y[:N_train]
print(X_train.shape, y_train.shape)

X_test = X[N_train:N_train+N_test,:]
y_test = y[N_train:N_train+N_test]
print(X_test.shape, y_test.shape)


Ns_train = 10*52 - look_back # cantidad de instancias que vamos a usar para entrenar. 
Ns_test = 4*52 #cantidad de puntos que vamos a usar para evaluar
Ns_total = Ns_train + Ns_test

length_totals = Ns_train + Ns_test + look_back

Xs_train = Xs[:Ns_train,:]
ys_train = ys[:Ns_train]
print(Xs_train.shape, ys_train.shape)

Xs_test = Xs[Ns_train:Ns_train+Ns_test,:]
ys_test = ys[Ns_train:Ns_train+Ns_test]

N1_train = 10*52 - look_back # cantidad de instancias que vamos a usar para entrenar. 
N1_test = 4*52 #cantidad de puntos que vamos a usar para evaluar
N1_total = N1_train + N1_test

length_total1 = N1_train + N1_test + look_back

X1_train = X1[:N1_train,:]
y1_train = y1[:N1_train]
print(X1_train.shape, y1_train.shape)

X1_test = X1[N1_train:N1_train+N1_test,:]
y1_test = y1[N1_train:N1_train+N1_test]

Ne_train = 10*52 - look_back # cantidad de instancias que vamos a usar para entrenar. 
Ne_test = 4*52 #cantidad de puntos que vamos a usar para evaluar
Ne_total = Ne_train + Ne_test

length_totale = Ne_train + Ne_test + look_back

Xe_train = Xe[:Ne_train,:]
ye_train = ye[:Ne_train]
print(Xe_train.shape, ye_train.shape)

Xe_test = Xe[Ne_train:Ne_train+Ne_test,:]
ye_test = ye[Ne_train:Ne_train+Ne_test]
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs = 4).fit(X_train,y_train)
rfS = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs = 4).fit(Xs_train,ys_train)
rf1 = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs = 4).fit(X1_train,y1_train)
rfe = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs = 4).fit(Xe_train,ye_train)
```

**Optimización de hiperparametros**

```python
# Definimos las listas vacias para los valores de accuracy deseados
lista_accuracy_train = []
lista_accuracy_test = []

from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor


# Definimos la lista de valores de k que vamos a explorar
k_vecinos = [1,5,10,20,30,40,50,70,85,100,150,200]

# Generamos en loop sobre los distintos valores de k 
for k in k_vecinos:
    
    # Vamos a repetir el siguiente bloque de código
    
    # Definir el modelo con el valor de vecinos deseado
    rfOP = RandomForestRegressor(n_estimators=k, random_state=42, n_jobs = 4)
    
    # Entrenar el modelo
    rfOP.fit(X_train,y_train)
    
    # Predecir y evaluar sobre el set de entrenamiento
    y_train_pred = rfOP.predict(X_train)
    y_test_pred = rfOP.predict(X_test)
    
    # Errores
    
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    # Agregar la información a las listas
    lista_accuracy_train.append(rmse_train)
    lista_accuracy_test.append(rmse_test)
```

### 2.2. Multi-Step Prediction

Con el modelo ya entrenado, vamos a usar el método recursivo para hacer Multi-Step prediction.

```python
y_ms_test_pred = []
x = X_test[0,:] # arrancamos con la primera instancias del conjunto de test
for i in range(y_test.size):
    y_os_pred = rf.predict(x.reshape(1,-1))
    y_ms_test_pred.append(y_os_pred)
    x = np.append(x[1:], y_os_pred)
y_ms_test_pred = np.array(y_ms_test_pred)
```
